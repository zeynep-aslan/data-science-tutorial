## Veri Seti Hikayesi

```python
import seaborn as sns  # görece diğerlerine göre daha zor problemlerde kullanırız bu grafigi
fmri = sns.load_dataset("fmri")  # beyne bağlanan bir cihaz aracılığıyla toplanan sinyalleri ifade eden bir veri seti
df = fmri.copy()
df.head()
# zamana bağlı d.lerin old.veri setlerinde ve iot cihazların ürettiği verileri görselleşt.için
# subject: verilerin toplandığı kişiler, timepoint: zaman noktaları, event: verinin toplanmasıyla ilgili olaylar
# region: sinyalin toplandığı bölge, signal: gelen sinyal
```

```python
df.shape
```

```python
df["timepoint"].describe()  # timepoint: kesikli bir sayısal d. kategorik d.mişcesine muamele görebilecek bir d.
```

```python
df["signal"].describe()  # standart sapma yüksek, sayı değerleri çok yüksek değil, bu da bize mekanik bir veri olabileceğinin izlenimini veriyor
```

```python
df.groupby("timepoint")["signal"].count()  # elimize böyle veri setleri geldiğinde aklımıza gelen ilk şey tekillik olur, unique lik olur
# subject ler unique mi tekil mi çoklamış mı her biri için ayrı timepoint mi var
# her bir zaman noktasında eşit sinyal mi var? sorusunun cevabı için groupby() yaparız
# timepoint d.ine göre signal i grupla .count() ile sinyalleri sayar
# çıktıya göre: demek ki her bir zaman noktasına göre eşit sayma işlemi yapılmış, her bir zaman noktasında eşit sinyal toplanmış
# yani mekanik bir veri seti
# her bir zaman noktasından 56 tane sinyal alındıysa alınan her bir sinyalin eşsiz(unique) olmasını bekleriz
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```
